```python
# Test the refactored get_dataset with automatic schema loading
import importlib
import brasa.queries
importlib.reload(brasa.queries)

from brasa.queries import get_template_schema, get_template_partitioning, get_dataset

# Test 1: Get schema from a template with fields defined
schema = get_template_schema("b3-futures-settlement-prices-pipeline")
print("Schema from template:")
print(schema)
print()

# Test 2: Get partitioning from template
partitioning = get_template_partitioning("b3-futures-settlement-prices-pipeline", schema)
print("Partitioning from template:")
print(partitioning)
print()

# Test 3: Get dataset with auto-schema (default behavior)
ds_with_schema = get_dataset("b3-futures-settlement-prices-pipeline")
print("Dataset schema (auto-loaded):")
print(ds_with_schema.schema)
print()

# Test 4: Get dataset without auto-schema
ds_without_schema = get_dataset("b3-futures-settlement-prices-pipeline", use_template_schema=False)
print("Dataset schema (from parquet):")
print(ds_without_schema.schema)
```

```python
# Test 4: Template without fields defined (ETL template)
# brasa-prices is an ETL dataset with no fields defined in template
schema_etl = get_template_schema("brasa-prices")
print(f"Schema from ETL template (no fields): {schema_etl}")

# Test 5: Non-existent template (returns None gracefully)
schema_nonexistent = get_template_schema("nonexistent-dataset")
print(f"Schema from non-existent template: {schema_nonexistent}")
```
